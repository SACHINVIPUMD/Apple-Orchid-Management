```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Data paths
leaf_dir = 'Apple Orchid Management/dataset2'
fruit_dir = 'Apple Orchid Management/dataset3'

# Data preprocessing for leaf and fruit classification
train_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_generator = train_datagen.flow_from_directory(
    directory='Apple Orchid Management/combined dataset',  # combine both datasets in a folder
    target_size=(128, 128),
    batch_size=32,
    class_mode='binary',  # Binary classification (Fruit or Leaf)
    subset='training'
)

val_generator = train_datagen.flow_from_directory(
    directory='Apple Orchid Management/combined dataset',
    target_size=(128, 128),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)

# Model architecture
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')  # Binary classification (Fruit or Leaf)
])

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(train_generator, epochs=10, validation_data=val_generator)

# Save the model
model.save('fruit_leaf_classifier.h5')
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Define dataset directory
leaf_dir = 'Apple Orchid Management/dataset2/train_augmented'

# Initialize ImageDataGenerator with validation split
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2  # 20% for validation
)

# Create training and validation data generators
leaf_train_gen = train_datagen.flow_from_directory(
    directory=leaf_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',  # One-hot encoding
    subset='training'
)

leaf_val_gen = train_datagen.flow_from_directory(
    directory=leaf_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)

# Print class indices to verify
print("Class indices:", leaf_train_gen.class_indices)

# Define the model
leaf_model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(4, activation='softmax')  # 4 classes
])

# Compile the model
leaf_model.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
leaf_history = leaf_model.fit(leaf_train_gen, epochs=10, validation_data=leaf_val_gen)

# Save the model
leaf_model.save('leaf_disease_classifier.h5')
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Define dataset directory for fruits
fruit_dir = 'Apple Orchid Management/combined dataset/dataset3/ds/Test'

# Initialize ImageDataGenerator with validation split
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2  # 20% for validation
)

# Create training and validation data generators
fruit_train_gen = train_datagen.flow_from_directory(
    directory=fruit_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',  # One-hot encoding
    subset='training'
)

fruit_val_gen = train_datagen.flow_from_directory(
    directory=fruit_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)

# Print class indices to verify
print("Class indices:", fruit_train_gen.class_indices)

# Define the fruit disease classification model
fruit_model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(4, activation='softmax')  # 4 classes: Normal_Apple, Scab_Apple, Rot_Apple, Blotch_Apple
])

# Compile the model
fruit_model.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
fruit_history = fruit_model.fit(fruit_train_gen, epochs=10, validation_data=fruit_val_gen)

# Save the model
fruit_model.save('fruit_disease_classifier.h5')
```

```python
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image

# Load the pre-trained models
fruit_leaf_model = load_model('fruit_leaf_classifier.h5')
leaf_disease_model = load_model('leaf_disease_classifier.h5')
fruit_disease_model = load_model('fruit_disease_classifier.h5')

# Function to predict leaf or fruit and then disease
def predict_image(img_path):
    img = image.load_img(img_path, target_size=(128, 128))
    img_array = image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    # Step 1: Check if it's a fruit or leaf
    pred_fruit_leaf = fruit_leaf_model.predict(img_array)
    
    if pred_fruit_leaf < 0.5:
        print("It's a leaf!")
        pred_leaf_disease = leaf_disease_model.predict(img_array)
        leaf_classes = ['Healthy', 'Apple Rust', 'Black Rot', 'Apple Scab']
        print(f'Leaf Disease: {leaf_classes[np.argmax(pred_leaf_disease)]}')
    else:
        print("It's a fruit!")
        pred_fruit_disease = fruit_disease_model.predict(img_array)
        fruit_classes = ['Blotch Apple', 'Normal Apple', 'Rot Apple', 'Scab Apple']
        print(f'Fruit Disease: {fruit_classes[np.argmax(pred_fruit_disease)]}')

# Example prediction
predict_image('apple/grayspot.jpg')
```
